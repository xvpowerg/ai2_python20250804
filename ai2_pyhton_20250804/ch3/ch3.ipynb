```python
import pandas as pd
df = pd.read_csv("tips.csv")
df.head()
```

```python
def last_four(num):
    return str(num)[-4:]
```

```python
newStr = last_four(3560325168603410)
print(newStr)
```

```python
df["CC Number"].apply(last_four)
```

```python
def yelp(price):
    if price <10:
        return "$"
    elif price >= 10 and price < 30:
        return "$$"
    else:
        return "$$$"
yelp(35)    
```

```python
df["total_bill"].mean()
```

```python
df["Expensive"] = df["total_bill"].apply(yelp)
df.head(20)
```

```python
df["total_bill"].apply(lambda bill:bill * 0.18).head(5)
```

```python
def quality(total_bill,tip):
    if tip / total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
    
print(quality(23,15))
```

```python
df["Tip Quality"] = df[["total_bill","tip"]].apply(lambda df:quality(df["total_bill"],df["tip"]),axis=1)
```

## df.corr() for correlation checks

[Wikipedia on Correlation](https://en.wikipedia.org/wiki/Correlation_and_dependence)

相關係數的含義：

1 表示完全正相關：當一個變量增加時，另一個變量也同樣比例增加

。
-1 表示完全負相關：當一個變量增加時，另一個變量同樣比例減

少。
0 表示沒有線性相關：兩個變量之間沒有線性

關聯。
